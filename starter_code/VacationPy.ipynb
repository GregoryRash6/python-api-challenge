```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import gmaps
import os
from pprint import pprint

# Import API key
from api_keys import g_key
```

### Store Part I results into DataFrame
* Load the csv exported in Part I to a DataFrame

```python
# Load CSV from Part One
weather_data = pd.read_csv("../output_data/weather.csv")
weather_data
```

### Humidity Heatmap
* Configure gmaps.
* Use the Lat and Lng as locations and Humidity as the weight.
* Add Heatmap layer to map.

```python
# Configure gmaps
gmaps.configure(api_key=g_key)
```

```python
# Set Location and Humidty for Heat Map
locations = weather_data[["Lat", "Lng"]]
humidity = weather_data["Humidity"].astype(float)
```

```python
# Set MAx Intensity
max_intensity = weather_data["Humidity"].max()
```

```python
# Plot Heatmap
fig = gmaps.figure(center=(25,0), zoom_level=1.75)

# Create heat layer
heat_layer = gmaps.heatmap_layer(locations, weights=humidity, max_intensity=max_intensity)


# Add layer
fig.add_layer(heat_layer)

# Display figure
fig
```

### Create new DataFrame fitting weather criteria
* Narrow down the cities to fit weather conditions.
* Drop any rows will null values.

```python
# Drill DataFrame by Temperature
ideal_temp = weather_data[weather_data["Max Temp"] > 70]
ideal_temp = ideal_temp[ideal_temp["Max Temp"] < 80]
ideal_temp
```

```python
# Drill DataFrame by Windiness
ideal_wind = ideal_temp[ideal_temp["Wind Speed"] < 10]
ideal_wind
```

```python
# Drill DataFrame by Cloud Coverage
ideal_destination = ideal_wind[ideal_wind["Cloudiness"] == 0]
ideal_destination
```

### Hotel Map
* Store into variable named `hotel_df`.
* Add a "Hotel Name" column to the DataFrame.
* Set parameters to search for hotels with 5000 meters.
* Hit the Google Places API for each city's coordinates.
* Store the first Hotel result into the DataFrame.
* Plot markers on top of the heatmap.

```python
# Copy New DataFrame
hotel_df = ideal_destination.copy()

# Reset and Drop Index
hotel_df = hotel_df.reset_index()
del hotel_df["index"]

# Add Hotel Name Column
hotel_df["Hotel Name"] = ""
hotel_df
```

```python
# Create variables to clean up loop
lat = hotel_df["Lat"]
long = hotel_df["Lng"]
range_x = len(hotel_df)
```

```python
# Create For Loop to find Hotels Nearby 
for x in range(range_x):
    base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"
    params = {
        "location": f"{lat[x]},{long[x]}", 
        "radius": 5000,
        "type": "lodging",
        "keyword": "Hotel",
        "key": g_key,
    }
    response = requests.get(base_url, params=params).json()
    
    try:
        # extract results
        results = response["results"][0]["name"]
        print(f"The closest hotel is {results}.")
        hotel_df.loc[hotel_df["City"] == hotel_df["City"][x], "Hotel Name"] = results        
    except (IndexError):
        print("No lodging options within 5000 meters... skipping.")
    print("----------")
```

```python
# Drop all rows without lodging options
hotel_df = hotel_df[hotel_df["Hotel Name"] != ""]

# Drop and Reset Index
hotel_df = hotel_df.reset_index()
del hotel_df["index"]
```

```python
# Display new DataFrame
hotel_df
```

```python
# NOTE: Do not change any of the code in this cell

# Using the template add the hotel marks to the heatmap
info_box_template = """
<dl>
<dt>Name</dt><dd>{Hotel Name}</dd>
<dt>City</dt><dd>{City}</dd>
<dt>Country</dt><dd>{Country}</dd>
</dl>
"""
# Store the DataFrame Row
# NOTE: be sure to update with your DataFrame name
hotel_info = [info_box_template.format(**row) for index, row in hotel_df.iterrows()]
locations = hotel_df[["Lat", "Lng"]]
```

```python
# Add marker layer ontop of heat map
markers = gmaps.marker_layer(locations, info_box_content=hotel_info)
fig.add_layer(markers)


# Display Map
fig
```
